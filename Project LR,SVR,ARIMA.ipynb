```python
import warnings
import pandas as pd
import numpy as np
import datetime
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
def predictSVM(data,company):
    data.index=pd.to_datetime(data['Date'])
    data=data.drop(['Date'],axis='columns')
    data=data.loc['01-01-2007':]

    #data=data.loc['01-01-2008':'12-31-2011','01-01-2020':'12-31-2020','01-01-2021':'12-31-2021']
    #print(data)

    X=data[['Open','Close']]
    y=data[['Adjusted Close']]
    X_Train,X_Test,y_Train,y_Test=train_test_split(X,y,test_size=0.3,random_state=None)

    model=SVR(kernel='rbf')
    model.fit(X_Train,y_Train)
    y_Test['Predicted Adjusted Close']=model.predict(X_Test)
    #print(X)
    print("Mean squared error of",company,mean_squared_error(y_Test['Adjusted Close'],y_Test['Predicted Adjusted Close']))

    ylabel=company+" Stock Price"
   
    y_Test[['Adjusted Close','Predicted Adjusted Close']].loc['01-01-2009':'12-31-2009'].plot(xlim=([datetime.date(2009, 1, 1), datetime.date(2009, 12, 31)]),title="Stock price during Swine flu",ylabel=ylabel,figsize=(8,5))
    
    y_Test[['Adjusted Close','Predicted Adjusted Close']].loc['01-01-2020':'12-31-2020'].plot(xlim=([datetime.date(2020, 1, 1), datetime.date(2020, 12, 31)]),title="Stock price during Covid-19 2020",ylabel=ylabel,figsize=(8,5))
    
    y_Test[['Adjusted Close','Predicted Adjusted Close']].loc['01-01-2021':'12-31-2021'].plot(xlim=([datetime.date(2021, 1, 1), datetime.date(2021, 12, 31)]),title="Stock price during Covid-19 2021",ylabel=ylabel,figsize=(8,5))
       
    plt.show()
    
    return y_Test['Predicted Adjusted Close']
```

```python
import math
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
def augmented_dickey_fuller_test(data):
    dftest = adfuller(data)  
    print(dftest)   
```

```python
def predictARIMA(df,company):
    print(company,"Prediction")
    df.index=pd.to_datetime(df['Date'])
    df=df.drop(['Date'],axis='columns')
    df=df.loc['01-01-2020':'12-31-2021']
    ts = df['Adjusted Close']
    ts.head()
    plt.rcParams['figure.figsize'] = 30, 10
    plt.plot(ts)
    # log transformation
    ts_log = np.log(ts)
    augmented_dickey_fuller_test(ts_log)
    
    
    train_data, test_data = ts_log[:int(len(ts_log)*0.9)], ts_log[int(len(ts_log)*0.9):]
    plt.figure(figsize=(10,6))
    title=company+" Prices"
    plt.title(title)
    plt.plot(ts_log, c='r', label='training dataset') 
    plt.plot(test_data, c='b', label='test dataset')
    plt.legend()
    
    # To get p,q values visualize PACF(Partial Autocorrelation), ACF(Autocorrelation) respectively 
    plot_acf(ts_log)  
    plot_pacf(ts_log)  
    plt.show()
    

    # Find the first difference
    diff_1 = ts_log.diff(periods=1).iloc[1:]
    diff_1.plot(title='Difference 1st')
    augmented_dickey_fuller_test(diff_1)
    # Find the first difference if needed
    diff_2 = diff_1.diff(periods=1).iloc[1:]
    diff_2.plot(title='Difference 2nd')
    augmented_dickey_fuller_test(diff_2)
    
    # Build Model        
    history = [x for x in train_data]
    predictions =[]
    for t in range(len(test_data)):
        model = ARIMA(history, order=(1,1,0))
        model_fit = model.fit(disp=0)
        predictions.append(model_fit.forecast()[0])
        history.append(test_data[t])
    error = mean_squared_error(test_data, predictions)
    print("Mean squared error of",company,error)
    
    #build model
    model = ARIMA(train_data, order=(1, 1, 0))  
    fitted_m = model.fit(disp=0) 
    fitted_m.plot_predict()  
    
```
